```python
import torch
from llm.create import get_send_fn
from search.searcher import Searcher

print(torch.cuda.is_available())
```

```python
### Prameters to play with ###
llm_source = "huggingface"
model_name = "meta-llama/Meta-Llama-3.1-8B-Instruct"
knowledge_graph_dir = "./index_results/atelectasis_hg2"
##############################

searcher = Searcher(
    input_dir=knowledge_graph_dir,
    send_to=get_send_fn(soure=llm_source, model_name=model_name),
    community_level=1,
)
```

```python
result = searcher.search(
    query="What are symptoms of atelectasis?",
)
```
