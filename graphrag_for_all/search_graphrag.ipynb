```python
import torch
from llm.create import get_send_fn
from search.searcher import Searcher

print(torch.cuda.is_available())
```

```python
### Prameters to play with ###

llm_source = "openai"
model_name = "gpt-3.5-turbo"
knowledge_graph_dir = "./index_results/graphrag/atelectasis_top_1_chatgpt3_5"


# llm_source = "huggingface"
# model_name = "meta-llama/Meta-Llama-3.1-8B-Instruct"
# knowledge_graph_dir = "./index_results/graphrag/atelectasis_top_1_llama3_1/"
##############################

searcher = Searcher(
    input_dir=knowledge_graph_dir,
    send_to=get_send_fn(source=llm_source, model_name=model_name),
    community_level=1,
)
```

```python
result = searcher.search(
    query="What are symptoms of atelectasis?",
)
```
